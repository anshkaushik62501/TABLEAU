# **Publishing and Sharing in Tableau**

**1.What is the key difference between publishing a dashboard on Tableau Public versus Tableau Server/
Cloud in terms of accessibility and data security?**

**Ans-**

**Tableau Public:**

- Accessibility: Publicly available to anyone on the internet
- Data Security: No data security; all data is visible

**Tableau Server / Cloud:**

- Accessibility: Restricted to authorized users only
* Data Security: Secure with authentication and access control

**2.When publishing a Tableau dashboard, what happens to a live connection compared to an extract?
How does this affect data refresh and performance?**

**ans-**

**Live Connection:**

- Connects directly to the source database
- Always shows real-time data (no manual refresh)
- Performance depends on database speed/network → can be slower

**Extract (Hyper):**

- Data is stored as a snapshot in Tableau
- Needs scheduled/manual refresh
- Faster performance (optimized for Tableau)

**Effect:**

- Live → real-time but slower
- Extract → fast but needs refresh

**3.In Tableau Server or Cloud (conceptually), which feature allows administrators to control which users
can view, interact with, or edit a published dashboard?**

**ans-**


In Tableau Server or Tableau Cloud, administrators control access to dashboards using Permissions.

Permissions are set at different levels such as project, workbook, or view, and they define what actions a user can perform, including:

- Viewing a dashboard
- Interacting with filters or data
- Editing or publishing content

Permissions are assigned to users or groups and are based on predefined capabilities (like View, Explore, Publish, etc.).

Additionally, user roles (such as Viewer, Explorer, Creator) determine the overall level of access a user has on the platform.

**4.If you want users to view a dashboard but not download the underlying data, which permission setting
would you adjust in Tableau Server**

**Ans-**

To allow users to view a dashboard but not download the underlying data, you need to adjust the Permissions.

Specifically:

Allow: View
Deny/Remove: Download Full Data and Download Summary Data

**6.List two best practices you should follow when preparing a dashboard for public sharing to ensure data
security and privacy**

**Ans-**


**Remove sensitive data:**


- Do not include personal, confidential, or private information before publishing.


**Use aggregated or anonymized data:**


- Show summarized data instead of detailed records to protect individual information.

**7.After publishing a dashboard, users report a “Data not found” error.
 What is the most common cause related to data source settings?**

 **Ans-**

 The most common cause is that the data source connection is not accessible to Tableau Server/Cloud.

This usually happens when:

* The dashboard is still connected to a local file (e.g., Excel on your computer)
* The data source was not published or embedded properly
* The server does not have permission/access to the database or file path

**8.When embedding a Tableau dashboard into a secure internal portal, what two additional security
measures should be implemented to ensure only authorized users can access the embedded content?

**Ans-**

User Authentication (SSO / Trusted Authentication):
Ensure users are authenticated before accessing the embedded dashboard (e.g., Single Sign-On).

Row-Level Security (RLS):
Restrict data so users can only see data relevant to them.
