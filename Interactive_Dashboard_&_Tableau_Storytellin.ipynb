# **Interactive Dashboard & Tableau Storytelling**

**Q1 Explain the difference between a Tableau Dashboard and a Tableau Story Point. Give two healthcare-related
use cases.**

**Ans-**

| Feature     | **Dashboard**                                   | **Story Point**                                     |
| ----------- | ----------------------------------------------- | --------------------------------------------------- |
| Definition  | Collection of multiple worksheets in one screen | Sequence of dashboards/visuals arranged like slides |
| Purpose     | Monitor & analyze data at one place             | Present a narrative or step-by-step insights        |
| Interaction | Highly interactive (filters, actions)           | Limited interaction, more presentation-focused      |
| Structure   | Single view                                     | Multiple views connected as a story                 |
| Usage       | Real-time analysis                              | Explaining trends or findings                       |

**Q2 What are Actions in Tableau? Explain how Filter Action can help hospital management analyze patient data**

**Ans-**

Actions are interactive features that let one visualization affect another.
They create dynamic linking between charts, dashboards, and data.

👉 Types include:

* Filter Action
- Highlight Action
- URL Action


**Filter Action (Concept)**

**A Filter Action allows users to click/select data in one chart and filter related data in other charts.**


Use in Hospital Management

Example:

Dashboard has:
Chart 1 → Department (Cardiology, Neurology, etc.)
Chart 2 → Patient details / Treatment cost

👉 When user clicks Cardiology:

Chart 2 updates to show only cardiology patients
