**Q8. What is the difference between a Worksheet and a Dashboard in Tableau?
Why is a dashboard more useful for decision-makers**

**Ans-**

| Aspect        | Worksheet                          | Dashboard                           |
| ------------- | ---------------------------------- | ----------------------------------- |
| Definition    | Single visualization (chart/graph) | Combination of multiple worksheets  |
| Purpose       | Analyze one view of data           | Provide **complete overview**       |
| Interactivity | Limited                            | High (filters, actions, drill-down) |
| Usage         | Data exploration                   | Presentation & decision-making      |


Why Dashboard is More Useful 🎯

Combines multiple insights in one screen

Enables real-time filtering & comparison

Helps identify patterns, trends, KPIs quickly

Supports faster and better decisions

**Q9.  What is Tableau Prep, and why is it important in real-world data projects?
What is Tableau Prep, and why is it important in real-world data projects?
Then explain how you would:
Remove null values
Remove duplicate records
Change incorrect data types (Answer in steps, no tool access required.)**

**Ans-**

What is Tableau Prep? 🧩

Tableau Prep is a data preparation tool used to clean, combine, and transform raw data before analysis in Tableau.

Why it’s Important in Real Projects 🚀

Ensures data quality & accuracy

Handles messy/raw data (nulls, duplicates, errors)

Improves analysis reliability

Saves time via visual data cleaning flows

Steps in Tableau Prep
1. Remove Null Values ❌

Open data → go to Profile Pane

Click field (e.g., Sales/State)

Click Null values → choose Exclude
(or use filter → remove null)



2. Remove Duplicate Records 🔁

Go to Clean Step

Select all columns (or key columns)

Click More Options (⋮) → Remove Duplicates



3. Change Incorrect Data Types 🔄

In Profile Pane, click data type icon (ABC / # / calendar)

Select correct type:

String → Number

Number → Date, etc.

Verify changes in data grid
